```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
voltage = np.linspace(0, 100, 50)
resistance = 30
```

 - V = I*R

```python
current_1 = voltage / resistance
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.legend()
plt.show()
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.xlabel('Voltage')
plt.ylabel('Current')
plt.legend()
plt.show()
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.xlabel('Voltage')
plt.ylabel('Current')

plt.title("Ohm's Law")
plt.legend()

plt.show()
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.xlabel('Voltage')
plt.ylabel('Current')

plt.xlim(20, 60)
plt.ylim(1, 3)

plt.title("Ohm's Law")
plt.legend()

plt.show()
```

```python
current_2 = voltage / resistance ** 2
```

```python
plt.plot(voltage, current_1, label='V=I*R')
plt.plot(voltage, current_2, label='V=I*R^2')

plt.xlabel('Voltage')
plt.ylabel('Current')

plt.legend()

plt.title("Results of an Experiment")

plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y, label='chart 1')
plt.title("y = log(x)")
plt.xlabel("X")
plt.ylabel("log(x)")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, np.exp(y), label='chart 2')
plt.title("y = x")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.plot(x, y, label='chart 1')
plt.title("y = log(x)")
plt.xlabel("X")
plt.ylabel("log(x)")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, np.exp(y), label='chart 2')
plt.title("y = x")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.subplot(2, 2, (3, 4))
plt.plot(x, np.sqrt(x), label='chart 2')
plt.title("y = sqrt(x)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.tight_layout()
plt.show()
```

### Box Plot

### Histogram

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(df.expenses, bins=100)
plt.title('Expenses')

plt.subplot(1, 2, 2)
plt.hist(df.bmi, bins=50)
plt.title('BMI')

plt.show()
```
