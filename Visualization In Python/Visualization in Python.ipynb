```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create the dataframe itself using the json file
data = pd.read_json(r'rain.json').round(3)
data
```

```python
print("df statistics: ")
data.describe().round(3)
```

```python
data.describe(include='all').round(3)
```

```python
data.plot(x='Month', y='Temperature', label='Temp.')
data.plot(x='Month', y='Rainfall')
plt.show()
```

```python
# Task 2: MatplotLib PyPlot

#Temprature
plt.figure(figsize= (15,5))
plt.plot( data['Month'], data['Temperature'], label = 'Temperature')
plt.legend()
plt.show()
```

```python
#Rainfall BarGraph

plt.figure(figsize=(15,5)) 
plt.bar( data['Month'], data['Rainfall'], label = 'Rainfall')
plt.legend()
plt.show()
```

```python
#Both temperature and Rainfall

plt.figure(figsize=(15,5)) 
plt.plot( data['Month'], data['Rainfall'], label = 'Rainfall')
plt.plot( data['Month'], data['Temperature'], label = 'Temperature in C')
plt.legend()
plt.show()
```

```python
#Seaborn

df = pd.read_csv(r'tempYearly.csv')
df.head()
```

```python
sns.scatterplot(x = 'Year',
               y = 'Temperature',
               data = df)
plt.show()
```

```python
#Regression line
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x = 'Year',
               y = 'Temperature',
               data = df)
plt.show()
```

```python
#Regression line with rainfall and temp

sns.regplot(x = 'Rainfall',
               y = 'Temperature',
               data = df)
plt.show()
```

```python
#HeatMap

data = pd.read_csv(r'birthYearly.csv')
data.head(20)
```

```python
#Pivot table

data_table = data.pivot("month", "year", "births")
data_table
```

```python
sns.heatmap(data_table,  annot = True, fmt = "d" )
plt.show()
```

```python
#Joint Plot

data = pd.read_csv(r'tempYearly.csv')
data.head()
```

```python
sns.jointplot(data = data, x = "Temperature", y = "Rainfall")
plt.show()
```

```python
#Hexagram
sns.jointplot(data = data, x = "Temperature", y = "Rainfall", kind='hex')
plt.show()
```

```python
#Regression Line
sns.jointplot(data = data, x = "Temperature", y = "Rainfall", kind='reg')
plt.show()
```
